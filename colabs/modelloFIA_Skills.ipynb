```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Salary.csv')
data
```

```python
# Selezionare solo le colonne richieste
columns_to_keep = ["Job Title", "Gender", "Age", "Education Level", "Years of Experience", "Senior"]
data_cleaned = data[columns_to_keep]

# Visualizzare le prime righe del dataset pulito
data_cleaned.head()
```

```python
missing_values = data_cleaned.isnull().sum()
missing_values
```

```python
# Trasformazione della colonna 'Gender' in valori booleani: Maschio = 1, Femmina = 0
gender_mapping = {'Male': 1, 'Female': 0}
data_cleaned['Gender'] = data_cleaned['Gender'].map(gender_mapping)

# Verifica delle modifiche
data_cleaned.head()
```

```python
job_title_counts = data_cleaned['Job Title'].value_counts()
job_title_counts
```

```python
single_occurrence_job_titles = job_title_counts[job_title_counts == 1]
single_occurrence_job_titles
```

```python
# Eliminazione dei titoli di lavoro che compaiono solo una volta
#cleaned_data_reduced = data_cleaned[~data_cleaned['Job Title'].isin(single_occurrence_job_titles.index)]
cleaned_data_reduced = data_cleaned[data_cleaned['Job Title'].isin(job_title_counts[job_title_counts > 1].index)]

# Dimensioni del dataset dopo l'eliminazione
reduced_dataset_shape = cleaned_data_reduced.shape
reduced_dataset_shape
```

```python
job_title_counts_new = cleaned_data_reduced['Job Title'].value_counts()
```

```python
# Identificazione dei titoli di lavoro che compaiono meno di 20 volte nel dataset
less_than_20_occurrences = job_title_counts_new[job_title_counts_new < 20]
less_than_20_occurrences
```

```python
# Identificazione dei titoli di lavoro con meno di 10 occorrenze
titles_with_less_than_10_occurrences = job_title_counts_new[job_title_counts_new < 10].index

# Eliminazione delle righe con questi titoli di lavoro
cleaned_data_reduced_further = cleaned_data_reduced[~cleaned_data_reduced['Job Title'].isin(titles_with_less_than_10_occurrences)]

# Dimensioni del dataset dopo l'ulteriore eliminazione
further_reduced_dataset_shape = cleaned_data_reduced_further.shape
further_reduced_dataset_shape
```

```python
#libreria che permette di binarizzare la colonna skills
from sklearn.preprocessing import MultiLabelBinarizer

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np



# Estrazione e trasformazione delle competenze in un formato adatto per MultiLabelBinarizer
skills_series = cleaned_data_reduced_further['Skills'].str.split(', ').apply(lambda x: x if isinstance(x, list) else [])

# Inizializzazione e applicazione di MultiLabelBinarizer per trasformare le competenze in colonne binarie
mlb = MultiLabelBinarizer()
skills_encoded = mlb.fit_transform(skills_series)

# Creazione di un DataFrame per le competenze codificate l'attributo mlb.classes serve a dare i nomi alle colonne
skills_df = pd.DataFrame(skills_encoded, columns=mlb.classes_)

# Codifica della variabile target 'Job Title'
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(cleaned_data_reduced_further['Job Title'])

# Preparazione delle altre feature escludendo 'Skills'
X_other_features = cleaned_data_reduced_further.drop(columns=['Skills', 'Job Title'])

# Riunione delle nuove colonne delle competenze con le altre feature numeriche/categoriche, escludendo le righe con indice in eccesso
X_final = pd.concat([X_other_features.reset_index(drop=True), skills_df.iloc[:X_other_features.shape[0]].reset_index(drop=True)], axis=1)

# Controllo delle dimensioni per assicurarsi che corrispondano
assert X_final.shape[0] == X_other_features.shape[0], "Mismatch in row count"

# Divisione del dataset finale in set di training e test
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X_final, y, test_size=0.2, random_state=42)

# Inizializzazione e addestramento del modello di classificazione
randomTree_model = RandomForestClassifier(random_state=None)
decisionTree_model = DecisionTreeClassifier()

# Addestramento del modello con il dataset finale
randomTree_model.fit(X_train_final, y_train_final)
decisionTree_model.fit(X_train_final, y_train_final)

# Predizione sul set di test finale
y_pred_final_r = randomTree_model.predict(X_test_final)
y_pred_final_d = decisionTree_model.predict(X_test_final)

# Applicazione della cross-validation
# Nota: X_train_final e y_train_final sono stati preparati in precedenza
# Assicurati che il dataset sia completo e pronto per la cross-validation
scores_accuracy_r = cross_val_score(randomTree_model, X_train_final, y_train_final, cv=5, scoring='accuracy')
scores_precision_r = cross_val_score(randomTree_model, X_train_final, y_train_final, cv=5, scoring='precision_macro')
scores_recall_r = cross_val_score(randomTree_model, X_train_final, y_train_final, cv=5, scoring='recall_macro')

scores_accuracy_d = cross_val_score(decisionTree_model, X_train_final, y_train_final, cv=5, scoring='accuracy')
scores_precision_d = cross_val_score(decisionTree_model, X_train_final, y_train_final, cv=5, scoring='precision_macro')
scores_recall_d = cross_val_score(decisionTree_model, X_train_final, y_train_final, cv=5, scoring='recall_macro')

# Calcolo della media delle metriche di performance
mean_accuracy_r = np.mean(scores_accuracy_r)
mean_precision_r = np.mean(scores_precision_r)
mean_recall_r = np.mean(scores_recall_r)

mean_accuracy_d = np.mean(scores_accuracy_d)
mean_precision_d = np.mean(scores_precision_d)
mean_recall_d = np.mean(scores_recall_d)

print("Random tree classifier result")
print(mean_accuracy_r, mean_precision_r, mean_recall_r)

print("Decision tree result")
print(mean_accuracy_d, mean_precision_d, mean_recall_d)
```
